```python
def PerfectNumbers(N):
    hashSet = set([1]) 
    for i in range(2,int((N+1)**.5+1)):
        cur = i*i
        while cur<=N:
            hashSet.add(cur)
            cur *=i
    ans = 0
    hList = sorted(list(hashSet))[::-1]
#     print ('hList=',hList)
    res = set()
    for i in hList:
        for j in hList:
            if i+j<=N:
                res.add(i+j)
    return len(res)
    
#     for i in range(1,N+1):
#         for j in hList:
#             if j>i:
#                 continue
#             if j<i/2:
#                 break
#             if i-j in hashSet:
#                 ans+=1
#                 break
#     return ans

print (PerfectNumbers(1))
print (PerfectNumbers(2))
print (PerfectNumbers(3))
print (PerfectNumbers(5))
print (PerfectNumbers(10**6))


            
```
